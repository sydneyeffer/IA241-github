```python
# Rename the columns
df.rename(columns={
    'Eligible Bachelors': 'Name',
    'Age': 'Age',
    'Hometown': 'Hometown',
    'Occupation': 'Occupation',
    'Income (Estimated)': 'Income',
    'Height (Estimated)': 'Height',
    'Education Level ': 'Education_Level',
    'Humor Rating (1-4)': 'Humor',
    'Emotional Intelligence (1-10)': 'Emotional_Intelligence',
    'Physical Attractiveness (1-10)': 'Attractiveness',
}, inplace=True)

# Check the updated column names
df.columns.tolist()
```

```python
# Strip leading and trailing spaces from all column names
df.columns = df.columns.str.strip()

# Rename the columns
df.rename(columns={
    'Eligible Bachelors': 'Name',  # This column will represent the name of the bachelors
    'Age': 'Age',
    'Hometown': 'Hometown',  # You may not need to use this one, but if you plan to use it, keep it
    'Occupation': 'Occupation',  # Same here, depending on if you plan to use it
    'Income (Estimated)': 'Income',  # Standardize this to just 'Income'
    'Height (Estimated)': 'Height',  # Standardize this to 'Height'
    'Education Level': 'Education_Level',  # Removed space at the end
    'Humor Rating (1-4)': 'Humor',  # Rename humor column for consistency
    'Emotional Intelligence (1-10)': 'Emotional_Intelligence',  # Cleaned up column name
    'Physical Attractiveness (1-10)': 'Attractiveness',  # Renamed correctly
}, inplace=True)

# Check the updated column names
df.columns.tolist()
```

```python
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
df['Height'] = pd.to_numeric(df['Height'], errors='coerce')
```

```python
preferences = {
    "Age": 29,
    "Income": 100000,
    "Height": 6.1,
    "Humor": 9,
    "Emotional_Intelligence": 9,
    "Attractiveness": 9,
    "Emotional_Intelligence":9,
    "Attractiveness":9
}
```

```python
weights = {
    "Age": 1.0,
    "Income": 1.5,
    "Height": 1.0,
    "Humor": 2.0,
    "Emotional_Intelligence": 2.0,
    "Attractiveness": 2.0
}

def score_bachelor(row):
    score = 0
    for trait in preferences:
        if preferences[trait] is not None:
            try:
                diff = abs(float(preferences[trait]) - float(row[trait]))
            except (TypeError, ValueError):
                diff = 0  # Handle errors by setting diff to 0

            if trait == "Income":
                norm_diff = diff / 100000  # Normalize Income for scale
            elif trait == "Height":
                norm_diff = diff / 12  # Normalize height for scale
            else:
                norm_diff = diff / 10  # Default normalization
            score += weights[trait] * (1 - norm_diff)
    return round(score * 10, 2)

# Apply the scoring function
df["Compatibility_Score"] = df.apply(score_bachelor, axis=1)
```

```python
top_matches = df.sort_values(by="Compatibility_Score", ascending=False).head(5)
top_matches[["Name", "Compatibility_Score"]]
```

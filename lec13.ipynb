```python
import pandas as pd 

df = pd.read_excel('s3://effer-241-2023-python/house_price.xls')
df['built_in'] = pd.to_datetime(df['built_in'],  format='%Y')
df.head()
```

```python
df.plot(x = 'built_in', y = 'price')
```

```python
avg_price_year = df.groupby('built_in')['price'].mean()

avg_price_year
```

```python
avg_price_year = df.groupby('built_in')['price'].mean()

avg_price_year.plot()
```
